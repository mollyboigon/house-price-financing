```python
import pandas as pd
```

I import the excel file, which I retrieved from this link https://www.census.gov/construction/nrs/historical_data/index.html

```python
df = pd.read_excel("data/pricefinc_cust.xls", skiprows = 5)
df
```

```python
df.head(5)
```

```python
df = df.rename(columns={"Unnamed: 0": "year",\
                        "Unnamed: 1": "med-price-total",\
                        "tional1": "med-price-conv",\
                        "Insured": "med-price-fha",\
                        "Guaranteed": "med-price-va",\
                        "Hous. Serv.": "med-price-rur",\
                        "Unnamed: 6": "med-price-cash",\
                        "Unnamed: 7": "av-price-total",\
                        "tional1.1": "av-price-conv",\
                        "Insured.1": "av-price-fha",\
                        "Guaranteed.1": "av-price-va",\
                        "Hous. Serv..1": "av-price-rur",\
                        "Unnamed: 12": "av-price-cash"})
```

```python
df
```

```python
df = df.drop(labels=[34,35], axis=0)
```

```python
df
```

```python
df.sort_values("av-price-cash", ascending=False)
```

```python
melt_df = pd.melt(df, id_vars = "year")
melt_df
```

```python
melt_df.value_counts("year")
```

```python
melt_df = melt_df.rename(columns={"variable": "loan-type",\
                                  "value" : "dollar-amt"})
melt_df
```

```python
type("dollar-amt")
```

```python
melt_df["dollar-amt"] = pd.to_numeric(melt_df["dollar-amt"], errors = "coerce")
melt_df
```
